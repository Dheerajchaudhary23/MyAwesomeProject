```python
import matplotlib.pyplot as plt
```

```python
# selecting the column to create a histogram
column_to_visualize = 'SepalLengthCm'
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(df[column_to_visualize], bins=20, color='skyblue')
plt.title(f'Distribution of {column_to_visualize}')
plt.xlabel(column_to_visualize)
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show() # Display the histogram
```

```python
# Now Selecting the numeric columns for visualization
numeric_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
# Create histograms for each numeric column
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    plt.hist(df[column], bins=20, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout() # Adjust layout and display the histograms
    plt.show()
```
